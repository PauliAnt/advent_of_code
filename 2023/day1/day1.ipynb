```python

import re


with open('input.txt') as f:
    lines = f.readlines()

digit_dict = {'one':'o1e','two':'t2o','three':'th3ee','four':'f4ur','five':'f5ve','six':'s6x','seven':'se7en','eight':'e8ght','nine':'n9ne'}

digit_sum = 0
for orig_line in lines:
    line = orig_line
    for digit in digit_dict.keys():
        if digit in line:
            line = line.replace(digit,digit_dict[digit])
            
            
    digits = re.sub('\D', '',line)
    digit_sum += int(digits[0]+digits[-1])
    print(orig_line,int(digits[0]+digits[-1]))

digit_sum
#53510
```
