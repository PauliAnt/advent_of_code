```python
with open('input.txt') as file:
    lines = file.readlines()

card_values = {'A':14, 'K':13, 'Q':12, 'T':10, '9':9, '8':8, '7':7, '6':6, '5':5, '4':4, '3':3, '2':2 ,'J':1}

hands = {}

for ii,line in enumerate(lines):
    line = line.replace('\n','')
    hand = line.split(' ')[0]
    bid = line.split(' ')[1]

    hand_value = sum([card_values[x] * 10**(8-2*jj) for jj,x in enumerate(hand)]) # example QQQAJ equals 1212121114


    card_counts = {}
    while hand != '':
        card = hand[0]
        card_counts[card] = hand.count(card)
        hand = hand.replace(card,'')
    

    if 'J' in card_counts.keys():
        joker_count = card_counts['J']
        card_counts['J'] = 0
    else:
        joker_count = 0


    counts = list(card_counts.values())
    # 1 high card, 2 one pair, 3 two pairs, 4 three of the same, 5 full house, 6 four same, 7 five same
    
    max_count_ind = counts.index(max(counts))
    counts[max_count_ind] += joker_count

    combination_value = counts[max_count_ind]

    if max(counts) >= 4:
        combination_value += 2
    elif 2 in counts and 3 in counts:
        combination_value += 2
    elif max(counts) == 3:
        combination_value += 1
    elif counts.count(2) == 2:
        combination_value += 1

    hand_value += 10**10*combination_value
    hands[hand_value] = int(bid)

winnings = 0
for ii,hand in enumerate(sorted(hands.keys())):
    winnings += hands[hand]*(ii+1)

winnings
```
