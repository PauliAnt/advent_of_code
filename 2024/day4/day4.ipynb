```python
with open('input.txt') as file:
    data = file.read()
```

```python

import re
s = 0

# horizontal 
data_hor = data
data_hor += '\n' + data_hor[::-1] # add reverse 

match = re.findall(r'XMAS', data_hor) # find matches
s += len(match)


# vertical 
lines = data.split('\n')

data_ver = ''
for ii in range(len(lines[0])):
    for jj in range(len(lines)):
        data_ver += lines[jj][ii]
    data_ver += '\n'

data_ver += '\n' + data_ver[::-1] # add reverse 

match = re.findall(r'XMAS', data_ver) # find matches
s += len(match)

## diagonal analysis, more like I need to be diagnosed

# diagonal up left -> right down
# lower triangle
data_diag_1 = ''
for ii in range(len(lines)):
    jj = 0
    while (ii + jj) < len(lines):
        data_diag_1 += lines[ii+jj][jj]
        jj +=1
    data_diag_1 += '\n'
# upper triangle
for ii in range(1,len(lines[0])):
    jj = 0
    while (ii + jj) < len(lines):
        data_diag_1 += lines[jj][ii+jj]
        jj +=1
    data_diag_1 += '\n'


data_diag_1 += '\n' + data_diag_1[::-1]  # add reverse 
match = re.findall(r'XMAS', data_diag_1) # find matches
s += len(match)


# diagonal up right -> left down
# lower triangle
data_diag_2 = ''
N = len(lines[0])-1
for ii in range(len(lines)):
    jj = 0
    while (ii + jj) < len(lines):
        data_diag_2 += lines[ii+jj][N-jj]
        jj +=1
    data_diag_2 += '\n'
# upper triangle
for ii in range(1,len(lines[0])):
    jj = 0
    while ii+jj < len(lines[0]):
        data_diag_2 += lines[jj][N-(ii+jj)]
        jj +=1
    data_diag_2 += '\n'

data_diag_2 += '\n' + data_diag_2[::-1] # add reverse 

match = re.findall(r'XMAS', data_diag_2) # find matches 
s += len(match)

    
print(s)
```
