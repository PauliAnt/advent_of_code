```python
with open('input.txt') as file:
    data = file.read()

data = [int(num) for num in data]
```

```python
# Silver, clean and simple
# Read memory data into list. 0 means empty and n means file with id = (n - 1)
seq = []
for ii in range(len(data)):
    if ii % 2 == 0:
        seq.extend([(ii)//2+1] * data[ii])
    else:
        seq.extend([0] * data[ii])

# ii starts from the top and goes throught the list finding empty spaces, jj goes in reverse order finding files
ii,jj = 0, (len(seq) - 1)
while ii < jj:
    while seq[ii] != 0 and ii < jj:
        ii += 1
    while seq[jj] == 0 and ii < jj:
        jj -= 1

    seq[ii], seq[jj] = seq[jj], seq[ii] # Sneaky oneliner to change positions for ii and jj
    


s = sum([ii * (seq[ii]-1) * (seq[ii] != 0) for ii in range(len(seq))])
s
```

```python
# Gold, "clean" and "simple"
seq = []
for ii in range(len(data)):
    if ii % 2 == 0:
        seq.extend([(ii)//2+1] * data[ii])
    else:
        seq.extend([0] * data[ii])


# Find file blocks and memory blocks
mem_blocks = []
file_blocks = []
ii = 0
while ii < len(seq):
    while ii < len(seq) and seq[ii] != 0:
        ii += 1
    mem_start = ii
    while ii < len(seq) and seq[ii] == 0:
        ii += 1
    mem_stop = ii
    mem_blocks.append((mem_start,mem_stop))

ii = 0
while ii < len(seq):
    while ii < len(seq) and seq[ii] == 0:
        ii += 1
    file_start = ii
    file_id = seq[ii]
    while ii < len(seq) and seq[ii] == file_id:
        ii += 1
    file_end = ii
    file_blocks.append((file_start, file_end))
file_blocks = file_blocks[::-1] # Reverse the list to go through them in right order


def swap_subsets(file, mem_start):
    # Helper function to change position of subsets.
    # Returns the new start index for the empty spaces
    file_id = seq[file[0]]
    file_size = file[1] - file[0]
    for ii in range(file_size):
        seq[file[0] + ii] = 0
        seq[mem_start + ii] = file_id
    return mem_start + ii + 1


for file in file_blocks:
    for mem in mem_blocks:
        if file[0] < mem[0]: # If file position is before mem space move to the next file
            break
        file_size = file[1] - file[0]
        mem_size = mem[1] - mem[0]
        # Check whether file fits in to the memory
        if file_size <= mem_size:
            mem_temp = swap_subsets(file,mem[0])
            mem_blocks.remove(mem) # Remove empty memory block
            # If the whole memory block was not used, replace it with new smaller memory block
            if mem_temp < mem[1]: 
                mem_blocks.append((mem_temp,mem[1]))
                mem_blocks.sort()
            break

s = sum([ii * (seq[ii]-1) * (seq[ii] != 0) for ii in range(len(seq))])
s
```
