```python
with open('input.txt') as file:
    lines = file.readlines()

# Start by saving list of operator indexes and types (1 for mult, 0 for +)
operators = []
for ii, char in enumerate(lines[-1][:-1]):
    if char != ' ':
        operators.append((ii, char=='*'))



# Split the data by precalculated indeces
operator_indices = [x[0] for x in operators]
data = []
for line in lines[:-1]:
    data.append([line[:-1][i:j-1] for i,j in zip(operator_indices, operator_indices[1:]+[len(line)])]) # Absolute str shredder monster


N = len(operators)
M = len(data)

# Once again abusing python str notation
solution = 0
for ii in range(N):
    # Initialise list of empty strings
    row_numbers = ['' for x in range(len(data[jj][ii]))]
    
    # Add numbers from top down to each str in list
    for jj in range(M):
        for kk, number in enumerate(data[jj][ii]):
            row_numbers[kk] += number

    # Convert to int
    row_numbers = [int(x) for x in row_numbers]
    
    # Calculate mult/sum
    row = row_numbers[0]
    for number in row_numbers[1:]:
        if operators[ii][1]:
            row *= number
        else:
            row += number
    # Profit
    solution += row



solution
```
