```python
with open('input.txt') as file:
    data = file.read()
    data = data.split(',')

# Generator object for possible splits
def splitting_factors(n):
    for i in range(1, n//2+1):
        if n % i == 0:
            yield i

invalid = []
for id_pair in data:
    first_id, last_id = (int(x) for x in id_pair.split('-'))
    
    
    for id_value in range(first_id, last_id+1):
        id_value = str(id_value)

        N = len(id_value)

        
        # Loop throught possible number of block sizes
        for N_block in splitting_factors(N):
            pattern = id_value[:N_block]
            
            is_invalid = True
            
            # To count as invalid only one block size needs to go throught and we can ignore the rest
            for split in range(1, N // N_block):
                # Check if pattern fits the wanted, stop at first non match
                if pattern != id_value[split * N_block : (split + 1) * N_block]:
                    
                    is_invalid = False
                    break
                
            if is_invalid:
                break
        if is_invalid:
            invalid.append(int(id_value))
                
    
sum(invalid)
```
