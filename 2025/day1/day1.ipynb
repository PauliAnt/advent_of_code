```python
with open('input.txt') as file:
    lines = file.readlines()
    
position = 50
zeros = 0
for line in lines:
    direction = line[0]
    amount = int(line[1:-1])   

    if direction == 'L':
        position = (position - amount) % 100
    elif direction == 'R':
        position = (position + amount) % 100
    
    if position == 0:
        zeros += 1
    
    
zeros
```

```python
with open('input.txt') as file:
    lines = file.readlines()
    
position = 50
clicks = 0
for line in lines:
    direction = line[0]
    amount = int(line[1:-1])
    
    if direction == 'L':
        # When starting from 0 there is always extra click on left 
        if position == 0:
            clicks -= 1
    
        position = (position - amount)
        
        while position < 0:
            position += 100
            clicks += 1
        
        # Extra check when moving from like 50 -> 0 to add a click
        if position == 0:
            clicks += 1
            
    elif direction == 'R':
        position = (position + amount)
        while position >= 100:
            position -= 100
            clicks += 1


    
clicks
```
