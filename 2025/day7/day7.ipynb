```python
import functools

with open('input.txt') as file:
    lines = file.readlines()
    
# STR grid
data_grid = []
for line in lines[:-1]:
    data_grid.append(line[:-1])
    
# INT grid
N, M = len(data_grid), len(data_grid[0])
grid = [[0 for ii in range(N)] for jj in range(M)]

for ii in range(N):
    for jj in range(M):
        if data_grid[ii][jj] == '^':
            grid[ii][jj] = 1
        if data_grid[ii][jj] == 'S':
            start_pos = (ii, jj)


# Recursion + cached results <33 This just brutally calculates all paths but since the results are cached it is actually very fast
@functools.cache
def go_downward(pos):
    next_pos = (pos[0]+1, pos[1])
    
    if next_pos[0] >= len(grid):
        return 0
    elif not (0 <= next_pos[1] < len(grid[0])):
        return 0
    
    
    if grid[next_pos[0]][next_pos[1]]:
        s = 1
        s += go_downward((next_pos[0], next_pos[1]-1))
        s += go_downward((next_pos[0], next_pos[1]+1))
        
    else:
        s = go_downward(next_pos)
    return s

splits = go_downward(start_pos)

splits+1 # Add the
```
