```python


with open('input.txt') as file:
    data = file.read()
    
id_list, _ = data.split('\n\n')


# Form a list containing all possible id ranges
fresh_ids = []
for id_pair in id_list.split('\n'):
    start_id, end_id = (int(x) for x in id_pair.split('-'))
    fresh_ids.append((start_id, end_id))

# Sort the list based on the starting point
fresh_ids = sorted(fresh_ids, key = lambda x: x[0])

# Using pythons dynamic list, merge ranges that are overlapping choosing the highest end point.
idx = 0
while idx < len(fresh_ids):

    start_id, end_id = fresh_ids[idx]

    # Starting from the second lowest starting point check if the starting point is in range
    for id_pair in fresh_ids[idx+1:]:
        
        start_id_ii, end_id_ii = id_pair

        if start_id <= start_id_ii <= end_id:

            # If the end point of the range is greater than current -> update
            if end_id_ii > end_id:
                fresh_ids[idx] = (start_id, end_id_ii)
                end_id = end_id_ii
            
            # The range is merged to our range and it can be removed
            fresh_ids.remove(id_pair)
            
        # Since the list is sorted we can move to the next when encountering too big value
        else:
            break
            
    # When there is nothing to merge we move to the next. Since we have removed all merged this automatically moves to the lowest non merged range
    idx += 1


# At this point we are left with non-overlapping list
fresh_food = 0
for id_pair in fresh_ids:
    fresh_food += id_pair[1]-id_pair[0]+1
    
fresh_food

```
