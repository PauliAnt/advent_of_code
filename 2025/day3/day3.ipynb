```python
with open('input.txt') as file:
    lines = file.readlines()

sum_numbers = []
for line in lines:
    numbers = line[:-1]
    tens = max(numbers[:-1])
    for ii,number in enumerate(numbers):
        if number == tens:
            break
    ones = max(numbers[ii+1:])
    
    sum_numbers.append(int(tens) * 10 + int(ones))

sum(sum_numbers)
```

```python
with open('input.txt') as file:
    lines = file.readlines()

sum_numbers = []
for line in lines:
    numbers = line[:-1]
    
    joltage = ''
    ii = 0
    n = 11
    N = len(numbers)
    while n >= 0:

        next_num = max(numbers[ii:N-n])
        for jj,number in enumerate(numbers[ii:N-n]):
            if number == next_num:
                ii = ii+jj+1
                break
        
        # Abuse Python's cursed str handling
        joltage += next_num
        
        n -= 1
        
    sum_numbers.append(int(joltage))
    

sum(sum_numbers)
```
