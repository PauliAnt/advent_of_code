```python
with open('input.txt') as file:
    lines = file.readlines()

sum_numbers = []
for line in lines:
    numbers = line[:-1]
    idx = numbers.index(max(numbers[:-1]))
    ones = max(numbers[idx+1:])
    
    sum_numbers.append(int(numbers[idx]) * 10 + int(ones))

sum(sum_numbers)
```

```python
with open('input.txt') as file:
    lines = file.readlines()

sum_numbers = []
for line in lines:
    numbers = line[:-1]
    
    joltage = ''
    idx = 0
    n = 11
    N = len(numbers)
    while n >= 0:
        subset = numbers[idx : N-n]
        next_num = max(subset)
        next_num_idx = subset.index(next_num)

        idx += next_num_idx+1

        
        # Abuse Python's cursed str handling
        joltage += subset[next_num_idx]
        
        n -= 1
        
    sum_numbers.append(int(joltage))
    

sum(sum_numbers)
```
